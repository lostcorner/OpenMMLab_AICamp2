```python
from mmpretrain.apis import ImageClassificationInferencer
from PIL import Image
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"]=["SimHei"] #设置字体
plt.rcParams["axes.unicode_minus"]=False #该语句解决图像中的“-”负号的乱码问题

```

```python
inference = ImageClassificationInferencer("resnet50_finetune.py", pretrained="./exp/best_accuracy_top1_epoch_42.pth")
img_path = 'apple.jpeg'
result = inference(img_path, show = True)

print(result)

```

```python
pred_label=result[0]['pred_label']
pred_score=result[0]['pred_score']
pred_class=result[0]['pred_class']
print(pred_label,pred_score,pred_class)
```

```python
img = Image.open(img_path)
plt.figure("ResNet50-fruit")  # 图像窗口名称
plt.imshow(img)
plt.axis('on')  # 关掉坐标轴为 off
plt.title("")  # 图像题目
plt.text(x=0,  # 文本x轴坐标
        y=0,  # 文本y轴坐标
        s='label:%d score:%f class:%s' % (pred_label,pred_score,pred_class),  # 文本内容
        ha='left',  # x=2.2是文字的左端位置，可选'center', 'right', 'left'
        va='bottom',  # y=8是文字的低端位置，可选'center', 'top', 'bottom', 'baseline', 'center_baseline'
        fontdict=dict(fontsize=8, color='black',
                    weight='bold',  # 磅值，可选'light', 'normal', 'medium', 'semibold', 'bold', 'heavy', 'black'

                    )  # 字体属性设置
        )
plt.show()

print(result)
```
